# Phase 1: Understanding the Problem

### **Summary of the Potential Business Impact of Delivery Delays**

Delivery delays can significantly affect a business's performance across multiple dimensions:

1. **Customer Satisfaction and Retention**

   * Delayed deliveries can frustrate customers, leading to decreased satisfaction, negative reviews, and loss of trust.
   * Repeat purchases and brand loyalty often decline when customers perceive the service as unreliable.

2. **Reputation Damage**

   * In the age of online reviews and social media, delivery issues can quickly damage a companyâ€™s public image.
   * Poor delivery performance can weaken the brand and reduce competitiveness in the market.

3. **Increased Operational Costs**

   * Delays may lead to higher costs due to re-delivery attempts, customer service interventions, and possible refunds or compensations.
   * Businesses might also need to invest more in logistics or expedite shipping to recover from delays.

4. **Supply Chain Disruption**

   * Consistent delivery issues may indicate inefficiencies in the supply chain, leading to stockouts, overstocking, or unbalanced inventory distribution.
   * This can impact production planning and vendor relationships.

5. **Loss of Revenue**

   * Unhappy customers are less likely to complete future transactions, directly affecting sales.
   * Delays can also impact large B2B clients who depend on timely delivery for their operations, risking contract loss.

6. **Legal and Contractual Penalties**

   * For businesses under strict service-level agreements (SLAs), frequent delays may result in penalties or legal action.
   * This is especially critical in industries like healthcare, construction, and retail.

7. **Competitive Disadvantage**

   * In a competitive market, companies with faster and more reliable delivery gain an edge.
   * Businesses suffering from frequent delays risk falling behind more efficient competitors.

### Conclusion:

Reducing delivery delays is not just a logistical concern but a strategic priority. It directly affects customer loyalty, operational efficiency, and long-term profitability. Businesses that proactively manage and improve their delivery systems can enhance their reputation and sustain growth in a competitive market.

### Dealing with outliers

Visualize the distribution of the critical numerical columns using box plots to identify outliers.


```python
fig, axes = plt.subplots(nrows=5, ncols=3, figsize=(18, 20))
axes = axes.flatten()

for i, col in enumerate(critical_numerical_cols):
    sns.boxplot(y=df[col], ax=axes[i])
    axes[i].set_title(f'Box plot of {col}')
    axes[i].set_ylabel(col)

# Remove any unused subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```


Reviewing the box plots from the previous step, several columns like `profit_per_order`, `sales_per_customer`, `order_item_discount`, `order_item_product_price`, `order_item_profit_ratio`, `sales`, `order_item_total_amount`, `order_profit_per_order`, and `product_price` show significant outliers. These outliers could skew model training. Given the nature of these financial and quantity-related variables, capping the outliers at a certain percentile (e.g., 5th and 95th) would be a reasonable approach to retain the majority of the data while mitigating the impact of extreme values. Columns like `shipping_duration` and `items_per_order` also show some outliers but seem less severe, so we will cap those as well. The columns `order_item_discount_rate` and `order_item_quantity` appear to have fewer extreme outliers, and their ranges are more constrained, so we will not apply capping to these.



```python
# Define the columns to cap and the percentile thresholds
cols_to_cap = [
    'profit_per_order',
    'sales_per_customer',
    'order_item_discount',
    'order_item_product_price',
    'order_item_profit_ratio',
    'sales',
    'order_item_total_amount',
    'order_profit_per_order',
    'product_price',
    'shipping_duration',
    'effective_price',
    'items_per_order',
]

for col in cols_to_cap:
    # Calculate the 5th and 95th percentiles
    lower_bound = df[col].quantile(0.05)
    upper_bound = df[col].quantile(0.95)

    # Apply capping
    df[col] = np.where(df[col] < lower_bound, lower_bound, df[col])
    df[col] = np.where(df[col] > upper_bound, upper_bound, df[col])

# Display the descriptive statistics after capping to see the effect
print("Descriptive statistics after capping outliers:")
display(df[cols_to_cap].describe())
```
